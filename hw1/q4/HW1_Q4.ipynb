```python
rank = np.linalg.matrix_rank(A)
A_t = A.transpose()
P = np.matmul(A, A_t)
Q = np.matmul(A_t, A)
```

```python
eig_val_P, eig_vec_P = np.linalg.eig(P)
eig_vec_P = np.matrix(eig_vec_P)
print(type(eig_vec_P))
eig_val_Q, eig_vec_Q = np.linalg.eig(Q)
eig_vec_Q = np.matrix(eig_vec_Q)
eig_vec_Q_trans = np.transpose(eig_vec_Q)
print(type(eig_vec_Q_trans))
```
